```typescript
// 217. Contains Duplicate

export const containsDuplicate = (nums: number[]): boolean => {
	return nums.length !== new Set(nums).size;
};
```

```typescript
// 242. Valid Anagram

export const isAnagram = (s: string, t: string): boolean => {
	if (s.length !== t.length) return false;
	const cnt = Array.from({ length: 26 }, () => 0);
	for (let i = 0; i < s.length; i++) {
		cnt[s.charCodeAt(i) - 97]++;
		cnt[t.charCodeAt(i) - 97]--;
	}
	return cnt.every((freq) => freq === 0);
};
```

```typescript
// 1299. Replace Elements with Greatest Element on Right Side

export const replaceElements = (nums: number[]): number[] => {
	let curr_max = -1;

	for (let i = nums.length - 1; i >= 0; i--) {
		const t = nums[i];
		nums[i] = curr_max;
		curr_max = Math.max(curr_max, t);
	}
	return nums;
};
```

```typescript
// 392. Is Subsequence

export const isSubsequence = (s: string, t: string): boolean => {
	if (s.length > t.length) return false;
	if (s.length === t.length) return s === t;

	let ss = 0;
	let tt = 0;

	while (tt < t.length) {
		if (s[ss] === undefined) break;

		if (s[ss] === t[tt]) ss++;
		tt++;
	}

	return ss === s.length;
};
```

```typescript
// 2486. Append Characters to String to Make Subsequence

export const appendCharacters = (s: string, t: string): number => {
	let ss = 0;
	let tt = 0;

	while (ss < s.length) {
		if (t[tt] === undefined) break;
		if (s[ss] === t[tt]) tt++;
		ss++;
	}

	return t.length - tt;
};
```

```typescript
// 3110. Score of a String

export const scoreOfString = (s: string): number => {
	let score = 0;
	for (let i = 1; i < s.length; i++) {
		score += Math.abs(s.charCodeAt(i - 1) - s.charCodeAt(i));
	}
	return score;
};
```

```typescript
// 58. Length of Last Word

export const lengthOfLastWord = (s: string): number => {
	let ans = 0;

	for (let i = s.length - 1; i >= 0; i--) {
		const char = s[i];
		if (ans === 0 && char === " ") continue;
		else if (char === " ") break;
		else ans++;
	}

	return ans;
};
```

```typescript
// 2678. Number of Senior Citizens

export const countSeniors = (strs: string[]): number => {
	let seniors = 0;
	for (let i = 0; i < strs.length; i++) {
		if (strs[i][11].concat(strs[i][12]) > "60") seniors++;
	}
	return seniors;
};
```

```typescript
// 1. Two Sum

export const twoSum = (nums: number[], target: number): [number, number] => {
	const map = new Map<number, number>();
	for (let i = 0; i < nums.length; i++) {
		if (map.has(nums[i])) return [map.get(nums[i])!, i];
		map.set(target - nums[i], i);
	}
	return [-1, -1];
};
```
